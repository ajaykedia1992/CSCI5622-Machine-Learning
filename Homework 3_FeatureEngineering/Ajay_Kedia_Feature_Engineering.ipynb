```python
class LemmaTokenizer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.wnl = WordNetLemmatizer()

    def fit(self, examples):
        # return self and nothing else
        return self

    def transform(self, examples):
        return [self.wnl.lemmatize(w) for w in word_tokenize(examples)]


class POSCountTokenizer(object):
    def __call__(self, text):
        return [word_and_pos[1] for word_and_pos in pos_tag(word_tokenize(text))]


class SpoilerWordDetectionTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        self.subjectivity = Subjectivity_Load()
        X = np.zeros((len(examples), 1));
        ii = 0
        for sentence in examples:
            count = 0
            words = word_tokenize(sentence.lower())
            for word in words:
                if word in self.subjectivity[4]:
                    count += 1
            X[ii, :] = np.array(count)
            ii += 1

        return csr_matrix(X)


class POSWordAssociationTokenizer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 200));
        ii = -1
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            ii = ii + 1
            jj = 0
            for word_and_pos in words_and_pos_tags:
                tag = word_and_pos[0] + "-" + word_and_pos[1]
                X[ii, jj] = np.array([len(tag)])
                jj = jj + 1
        return csr_matrix(X)


class WordStemmerTokenizer(object):
    def __init__(self):
        self.word = PorterStemmer()

    def __call__(self, doc):
        word_stem = [self.word.stem(w) for w in word_tokenize(doc)]
        return word_stem


class TextLengthTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        features = np.zeros((len(examples), 1))
        for ii,ex in enumerate(examples):
            features[ii, 0] = len(ex)
        return csr_matrix(features)


class ItemSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key

    def fit(self, x, y=None):
        return self

    def transform(self, examples):
        return examples[self.key]


class GrammarAdjectiveTranformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1));
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            count = 0
            i = 0
            for word_and_pos in words_and_pos_tags:
                if word_and_pos[1] == 'JJ' or word_and_pos[1] == 'JJR' or word_and_pos[1] == 'JJS':
                    count += 1
            X[i, :] = np.array([count])
            i += 1
        return csr_matrix(X)


class GrammarNounTranformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1));
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            count = 0
            i = 0
            for word_and_pos in words_and_pos_tags:
                if word_and_pos[1] == 'NN' or word_and_pos[1] == 'NNS' or word_and_pos[1] == 'NNP' or word_and_pos[
                    1] == 'NNPS':
                    count += 1
            X[i, :] = np.array([count])
            i += 1
        return csr_matrix(X)


class POSWordCountTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        pass

    def transform(self, examples):
        all_possible_tags = ['CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'LS', 'MD', 'NN', 'NNP', 'NNPS',
                             'NNS', 'PDT', 'POS', 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO', 'UH', 'VB',
                             'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB', 'PUNCT']
        dict = {k: v for v, k in enumerate(all_possible_tags)}
        X = np.zeros((len(examples), len(all_possible_tags)))
        ii = 0
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            for word_and_pos in words_and_pos_tags:
                tag = word_and_pos[1]
                if tag not in all_possible_tags:
                    tag = 'PUNCT'
                X[ii, dict[tag]] = np.array(len(tag))
            ii = ii + 1
        csr_matrix(X)


class GrammarSentenceTranformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        dict = {}
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            for word_and_pos in words_and_pos_tags:
                if word_and_pos[1] in dict:
                    dict[word_and_pos[1]] += 1
                else:
                    dict[word_and_pos[1]] = 1

        X = np.zeros((len(dict), 1));
        i = 0
        for key, value in dict.items():
            X[i, :] = np.array([value])
            i += 1
        return csr_matrix(X)


class BadWordTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        # with open("my_badlist.txt") as f:
        #     badwords = [l.strip() for l in f.readlines()]
        self.badwords_list = my_badlist

    def get_feature_names(self):
        return np.array(['spaces', 'capsratio', 'n_words', 'n_chars', 'allcaps', 'max_len', 'mean_len', 'bad_ratio', 'n_bad', '@','!'])

    def fit(self, examples):
        return self

    def transform(self, examples):
        # counting no of characters
        no_of_characters = [len(c) for c in examples]
        # counting no of words
        no_of_words = [len(c.split()) for c in examples]
        # number of uppercase words
        all_caps_words = [np.sum([w.isupper() for w in comment.split()]) for comment in examples]

        # average word length
        mean_length_of_words = [np.mean([len(w) for w in c.split()]) for c in examples]
        # number of google badwords:
        no_of_bad_words = [np.sum([c.lower().count(w) for w in self.badwords_list]) for c in examples]
        # longest word
        max_word_lenth = [np.max([len(w) for w in c.split()]) for c in examples]
        # count spaces
        spaces = [c.count(" ") for c in examples]
        # exclamation sign
        exclamation_sign_count = [c.count("!") for c in examples]
        # @ counts
        addressing_sign_count = [c.count("@") for c in examples]
        # all capital letter ratio
        all_caps_word_ratio = np.array(all_caps_words) / np.array(no_of_words, dtype=np.float)
        # bad word ration
        bad_word_ratio = np.array(no_of_bad_words) / np.array(no_of_words, dtype=np.float)

        return np.array([no_of_words, no_of_characters, all_caps_words, max_word_lenth, mean_length_of_words,
                         exclamation_sign_count, addressing_sign_count, spaces, bad_word_ratio, no_of_bad_words,
                         all_caps_word_ratio]).T


class RatingMovieTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1))
        ii = 0
        for x in examples:
            X[ii, :] = x
            ii += 1
        return csr_matrix(X)


class UserVoteTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1))
        ii = 0
        for x in examples:
            X[ii, :] = x
            ii += 1
        return csr_matrix(X)



class CountryTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        from collections import Counter
        country_dict = Counter(examples)
        X = np.zeros((len(examples), 1));
        for ii,x in enumerate(examples):
            if x in country_dict:
                X[ii,:] = country_dict[x]
            else:
                X[ii,:] = 0
        return csr_matrix(X)

class FirstAiredTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        from collections import Counter
        first_aired_dict = Counter(examples)
        X = np.zeros((len(examples), 1));
        for ii,x in enumerate(examples):
            if x in first_aired_dict:
                X[ii,:] = first_aired_dict[x]
            else:
                X[ii,:] = 0
        return csr_matrix(X)

class EpisodeTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1))
        ii = 0
        for x in examples:
            X[ii, :] = x
            ii += 1
        return csr_matrix(X)
    
    
class AllTextFeautresTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        # with open("my_badlist.txt") as f:
        #     badwords = [l.strip() for l in f.readlines()]
        self.badwords_list = my_badlist
        self.subjectivity = Subjectivity_Load()
        self.stemmer = nltk.stem.PorterStemmer()

    def get_feature_names(self):
        feature_list = []
        # feature_list.extend(self.pos_vect.get_feature_names())
        feature_list.extend(["n_nicks", "n_urls", "n_sentences","n_non_words", "idiot_regexp", "moron_regexp", 'n_words', 'n_chars', 'toolong', 'allcaps',
                             'max_len', 'mean_len', 'bad_ratio',"n_html", "strong_pos", "strong_neg","weak_pos","weak_neg", 'n_bad', 'capsratio'])
        feature_list.extend(["you_are_" + w for w in
                             self.you_word_related_tag_vector.get_feature_names()])
        feature_list.extend(self.unigram_feature_vectorization.get_feature_names())
        feature_list.extend(self.you_word_bigram_vectorization.get_feature_names())
        feature_list.extend(self.trigram_vect_you.get_feature_names())
        feature_list = [" ".join(w) if isinstance(w, tuple) else w
                        for w in feature_list]
        return np.array(feature_list)

    def fit(self, examples, y=None):
        self.fit_transform(examples, y)
        return self

    def fit_transform(self, examples, y=None):
        designed_words, filtered_words_lower, filtered_words, comments_prep = \
            self.feature_pre_process(examples)

        empty_analyzer = lambda x: x
        print("comment vecorizing")
        self.unigram_feature_vectorization = TfidfVectorizer(analyzer=empty_analyzer, min_df=3)
        
        unigrams = self.unigram_feature_vectorization.fit_transform(filtered_words_lower)
        
        self.you_word_related_tag_vector = TfidfVectorizer(token_pattern="(?i)you are(?: an?)?(?: the)?(?: as)? (\w+)")
        you_feature = self.you_word_related_tag_vector.fit_transform(comments_prep)
        
        collocation_you_bigram = col.BigramCollocationFinder.from_documents(filtered_words_lower)
        collocation_you_bigram.apply_freq_filter(3)
        
        collocation_you_bigram._apply_filter(lambda x, y: np.all([w != "you" for w in x]))
        
        #bigram measures
        self.bigram_measures = col.BigramAssocMeasures()
        # make tfidfvectorizer that uses these bigrams
        self.you_word_bigram_vectorization = TfidfVectorizer(analyzer=collocation_analyzer_for_words(collocation_you_bigram.nbest(self.bigram_measures.chi_sq, 1000)), min_df=3)
        bigram_words_for_you = self.you_word_bigram_vectorization.fit_transform(filtered_words_lower)
        # extract trigram collocations
        trigram_collocation_for_you = col.TrigramCollocationFinder.from_documents(filtered_words_lower)
        trigram_collocation_for_you.apply_freq_filter(3)
        trigram_collocation_for_you._apply_filter(lambda x, y: np.all([w != "you" for w in x]))
        self.trigram_words_for_you = trigram_collocation_for_you.nbest(col.TrigramAssocMeasures().chi_sq,1000)
        self.trigram_col_you = trigram_collocation_for_you
        self.trigram_vect_you = TfidfVectorizer(analyzer=collocation_analyzer_for_words(self.trigram_words_for_you, 3), min_df=3)
        trigram_words_for_you = self.trigram_vect_you.fit_transform(filtered_words_lower)
        ## some artificial features!
        designed_words.extend(self.artificial_features(filtered_words, examples,
                                          filtered_words_lower, ))
        designed_words = np.array(designed_words).T
        self.scaler = MinMaxScaler()
        designed_words = self.scaler.fit_transform(designed_words)
        features = []
        features.append(you_feature)
        # features.append(pos_unigrams)
        features.append(unigrams)
        features.append(bigram_words_for_you)
        features.append(trigram_words_for_you)
        features.append(sparse.csr_matrix(designed_words))
        features = sparse.hstack(features).tocsr()

        return features

    def get_comments_with_ascii(self,comments_without_nick_names,url_regex, comments_with_no_urls):
        comments_with_ascii = [c.replace(u'\xa0', ' ') for c in comments_with_no_urls]
        comments_with_ascii = [non_ascii_remove(c) for c in comments_with_ascii]
        comments_with_ascii = [c.replace("n't", "not").replace("N'T", "NOT").replace("'LL", "WILL").replace("'ll", "will") for c in
            comments_with_ascii]
        # replace /  with space, as this often separates words
        comments_with_ascii = [c.replace(u'/', ' ') for c in comments_with_ascii]
        ur = "you are "
        UR = "YOU ARE "
        comments_with_ascii = [re.sub(r"[Yy]ou'? ?a?re ", ur, c)for c in comments_with_ascii]
        # again for the loud people (don't want to lose that)
        comments_with_ascii = [re.sub(r"YOU'? ?A?RE ", UR, c)for c in comments_with_ascii]
        idiot = [len(re.findall("you.? [\w ]* idi.t", c))for c in comments_with_ascii]
        moron = [len(re.findall("you.? [\w ]* m.r.n", c))for c in comments_with_ascii]
        return comments_with_ascii,idiot,moron

    def filter_word_vectorization(self,flat_words):
        filtered_words = [[w for w in c
          if not re.findall(r"^[^\w]*$", w)] for c in flat_words]

        # get rid of non-word characters sourrounding words
        filtered_words = [[re.sub("^[^\w]*(\w+)[^\w]*$", r"\1", w) for w in c] for c in filtered_words]
        # laughter normalization ^^
        filtered_words = [[re.sub("(?i)ha(ha)+", r"haha", w) for w in c] for c in filtered_words]
        filtered_words = [[re.sub("(?i)l+o+l+(o+l+)+", r"lol", w) for w in c] for c in filtered_words]
        # replace the famous "0" as o
        filtered_words = [[re.sub("(?i)([a-z]+)0([a-z]+)", r"\1O\2", w) for w in c] for c in filtered_words]

        filtered_words = [[self.stemmer.stem(w) for w in c] for c in filtered_words]

        # detect weird stuff so we can spellcheck
        non_words = [[a for a in s if not nltk.corpus.wordnet.synsets(a)] for s in filtered_words]
        non_words = [[a for a in s if not a.lower() in self.badwords_list] for s in non_words]
        n_non_words = [len(w) for w in non_words]
        return filtered_words,n_non_words

    def feature_pre_process(self, comments):
        # remove nicknames, urls, html
        url_regex = re.compile(r"http[^\s]*")
        nick_regex = re.compile(r"@\w\w+:?")
        html_regex = re.compile(r"</?\w+[^>]*>")
        no_of_html_tag = [len(html_regex.findall(c)) for c in comments]
        comments = [html_regex.sub(' ', c) for c in comments]

        # count nickname in the comments
        no_of_nick_names_in_comments = [len(nick_regex.findall(c)) for c in comments]
        # comments without nicknames
        comments_without_nick_names = [nick_regex.sub('', c) for c in comments]

        no_of_urls_in_comments = [len(url_regex.findall(c)) for c in comments_without_nick_names]
        # replace with space
        comments_with_no_urls = [url_regex.sub(' ', c) for c in comments_without_nick_names]
        comments_with_ascii,idiot,moron =  self.get_comments_with_ascii(comments_without_nick_names,url_regex, comments_with_no_urls)

        # split into sentences
        sentences = [nltk.sent_tokenize(comment)
                     for comment in comments_with_ascii]
        # remove dots as they are annoying
        sentences = [[s.replace(".", " ") for s in sent] for sent in
                     sentences]
        # punctuation = \
        # ['...', '.', '?', '!', ',', "''", '``', '#', '$', "'", "%", "&"]
        no_of_sentences = [len(sent) for sent in sentences]
        words = [[nltk.word_tokenize(s) for s in sent] for sent in sentences]
        flat_words = [[w for sent in sents for w in sent] for sents in words]
        # remove "words" that contain no letter/numbers
        filtered_words, n_non_words = self.filter_word_vectorization(flat_words=flat_words)
        filtered_words_lower = [[w.lower() for w in comment]for comment in filtered_words]

        features = [no_of_nick_names_in_comments, no_of_urls_in_comments, no_of_sentences, n_non_words, idiot, moron,
                    no_of_html_tag]
        return [features, filtered_words_lower,
                filtered_words, comments_with_ascii]

    def get_subjectivity(self,filtered_words_lower, no_of_words):
        strong_pos = [np.sum([w in self.subjectivity[0] for w in c]) if len(c) else 0 for c in filtered_words_lower]
        strong_pos = np.array(strong_pos) / no_of_words
        strong_neg = [np.sum([w in self.subjectivity[1] for w in c]) if len(c) else 0 for c in filtered_words_lower]
        strong_neg = np.array(strong_pos) / no_of_words
        weak_pos = [np.sum([w in self.subjectivity[2] for w in c]) if len(c) else 0 for c in filtered_words_lower]
        weak_pos = np.array(strong_pos) / no_of_words
        weak_neg = [np.sum([w in self.subjectivity[3] for w in c]) if len(c) else 0 for c in filtered_words_lower]
        weak_neg = np.array(strong_pos) / no_of_words
        return strong_pos,strong_neg,weak_pos,weak_neg
        
    def artificial_features(self, filtered_words, comments, filtered_words_lower):
        ## some artificial features!
        no_of_words = np.array([len(c) for c in filtered_words], dtype=np.float)
        no_of_words += 0.1
        no_of_characters = [len(c) for c in comments]
        no_of_characters_too_long = np.array(no_of_characters) > 1000
        # number of uppercase words
        all_upper_case_words = [np.sum([w.isupper() for sentence in comment for w in sentence])for comment in filtered_words]
        # after removeing all the stuff above, the comment migh be empty
        maximum_word_length = [np.max([len(w) for w in c])if len(c) else 0 for c in filtered_words]
        # average word length
        mean_word_len = [np.mean([len(w) for w in c])if len(c) else 0 for c in filtered_words]

        no_of_bad_words = [np.sum([self.stemmer.stem(w) in self.badwords_list for w in c])if len(c) else 0 for c in filtered_words_lower]

        all_upper_caps_word_ratio = np.array(all_upper_case_words) / no_of_words
        
        bad_ratio = np.array(no_of_bad_words) / no_of_words

        # subjectivity database
        strong_pos,strong_neg,weak_pos,weak_neg = self.get_subjectivity(filtered_words_lower, no_of_words)

        result = [strong_pos, strong_neg, weak_pos, weak_neg, no_of_words, no_of_characters,
                  all_upper_case_words, no_of_characters_too_long, maximum_word_length, mean_word_len, bad_ratio,
                  no_of_bad_words, all_upper_caps_word_ratio]
        return result

    def transform(self, comments):
        designed_words, filtered_words_lower, filtered_words, comments_prep = \
            self.feature_pre_process(comments)
        unigrams = self.unigram_feature_vectorization.transform(filtered_words_lower)
        you_bigrams = self.you_word_bigram_vectorization.transform(filtered_words_lower)
        trigram_words_for_you = self.trigram_vect_you.transform(filtered_words_lower)
        you_words_in_comments_are = self.you_word_related_tag_vector.transform(comments_prep)
        designed_words.extend(self.artificial_features(filtered_words, comments,filtered_words_lower))
        designed_words = np.array(designed_words).T
        designed_words = self.scaler.transform(designed_words)
        features = []
        features.append(you_words_in_comments_are)
        # features.append(pos_unigrams)
        features.append(sparse.csr_matrix(designed_words))
        features.append(unigrams)
        features.append(you_bigrams)
        features.append(trigram_words_for_you)
        features = sparse.hstack(features).tocsr()
        return features

```

### Feature 5  : POS Tag Encoding

```python
import pandas as pd
from matplotlib import pyplot as plt
import nltk

all_possible_tags = ['WRB', 'WP$', 'WP', 'WDT', 'VBZ', 'VBP', 'VBN', 'VBG', 'VBD', 'VB', 'UH', 'TO', 'SYM', 'RP', 'RBS', 'RBR', 'RB', 'PRP$', 'PRP', 'POS', 'PDT', 'NNS', 'NNPS', 'NNP', 'NN', 'MD', 'LS', 'JJS', 'JJR', 'JJ', 'IN', 'FW', 'EX', 'DT', 'CD', 'CC']
all_possible_tags = sorted(all_possible_tags)
spoiler_pos_dict = {}
non_spoiler_pos_dict = {}

for tag in all_possible_tags:
    spoiler_pos_dict[tag] = 0
    non_spoiler_pos_dict[tag] = 0

dfTrain = pd.read_csv("../data/spoilers/train.csv")

for index,row in dfTrain.iterrows():
    sentence = row[kTEXT_FIELD]
    tags = nltk.pos_tag(sentence)

    spoiler = row[kTARGET_FIELD]
    d = None
    if spoiler:
        d = spoiler_pos_dict
    else:
        d = non_spoiler_pos_dict

    for tup in tags:
        if tup[-1] in d.keys():
            d[tup[-1]] += 1
        else:
            pass

print(spoiler_pos_dict)
print(non_spoiler_pos_dict)

D = spoiler_pos_dict

x = range(len(D))

y = []
z = []
for key,value in D.items():
    y.append(value)
    z.append(key)
plt.figure(figsize=(20,6))
plt.bar(x, y, align='center',color = 'red')
plt.xlabel('Part of Speech Tag')
plt.ylabel('Count of words')
plt.title('Spoiler : POS word Count')
plt.xticks(x, z)
plt.show()


y = []
z = []
D = non_spoiler_pos_dict
x = range(len(D))
for key,value in D.items():
    y.append(value)
    z.append(key)

plt.figure(figsize=(20,6))
plt.bar(x, y, align='center',color = 'black')
plt.xlabel('Part of Speech Tag')
plt.ylabel('Count of words')
plt.title('Non-Spoiler : POS word Count')
plt.xticks(x, z)

plt.show()
```

From the above graph, It's clear that pos tag is not playing much role as for the spoiler and non-spoiler graph, we are getting the almost same graph. So, this feature is not helping much.
