```python
def KeepWordsAndSpaces(text):
    punctuation_numbers_specialcharacters = re.compile(r'[^a-zA-Z ]')
    text = punctuation_numbers_specialcharacters.sub('', text)
    return text.lower()

def SentenceLength(text):
    text = KeepWordsAndSpaces(text)
    words = word_tokenize(text)
    yield len(words)
    
def make_collocation_analyzer(collocations, length=2):
    def analyzer(document):
        cols = [bigram for bigram in nltk.ngrams(document, length)
                    if bigram in collocations]
        return cols

    return analyzer

def remove_non_ascii(s):
    return "".join(i for i in s if ord(i) < 128)

def load_subjectivity():
    with open("load_subjectivity.pickle", 'rb') as fifty_f:
        subjectivity = pickle.load(fifty_f)
        return subjectivity
```

```python
class LemmaTokenizer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.wnl = WordNetLemmatizer()

    def fit(self, examples):
        # return self and nothing else
        return self

    def transform(self, examples):
        return [self.wnl.lemmatize(t) for t in word_tokenize(examples)]
class POSCountTokenizer(object):
    def __call__(self, text):
        words = word_tokenize(text)
        words_and_pos_tags = pos_tag(words)
        # Return a list of the part of speech for each word in the sentence.
        return [word_and_pos[1] for word_and_pos in words_and_pos_tags]
    
class SpoilerWordDetectionTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        self.subjectivity = load_subjectivity()
        X = np.zeros((len(examples), 1));
        ii = 0
        for sentence in examples:
            count = 0
            words = word_tokenize(sentence.lower())
            for word in words:
                if word in self.subjectivity[4]:
                    count += 1
            X[ii, :] = np.array(count)
            ii += 1

        return csr_matrix(X)
class POSWordAssociationTokenizer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 200));
        ii = -1
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            ii = ii + 1
            jj = 0
            for word_and_pos in words_and_pos_tags:
                tag = word_and_pos[0] + "-" + word_and_pos[1]
                X[ii, jj] = np.array([len(tag)])
                jj = jj + 1
        return csr_matrix(X)
    
class WordStemmerTokenizer(object):
    def __init__(self):
        self.word = PorterStemmer()

    def __call__(self, doc):
        word_stem = [self.word.stem(t) for t in word_tokenize(doc)]
        return word_stem


class TextLengthTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        features = np.zeros((len(examples), 1))
        i = 0
        for ex in examples:
            features[i, 0] = len(ex)
            i += 1

        return csr_matrix(features)


class ItemSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key

    def fit(self, x, y=None):
        return self

    def transform(self, examples):
        return examples[self.key]

class GrammarAdjectiveTranformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1));
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            count = 0
            i = 0
            for word_and_pos in words_and_pos_tags:
                if word_and_pos[1] == 'JJ' or word_and_pos[1] == 'JJR' or  word_and_pos[1] == 'JJS':
                    count += 1
            X[i, :] = np.array([count])
            i += 1
        return csr_matrix(X)

class GrammarNounTranformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1));
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            count = 0
            i = 0
            for word_and_pos in words_and_pos_tags:
                if word_and_pos[1] == 'NN' or word_and_pos[1] == 'NNS' or word_and_pos[1] == 'NNP' or word_and_pos[1] == 'NNPS':
                    count += 1
            X[i, :] = np.array([count])
            i += 1
        return csr_matrix(X)
    
class POSWordCountTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass
    def fit(self,examples):
        pass
    def transform(self,examples):
        all_possible_tags = ['CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'LS', 'MD', 'NN', 'NNP', 'NNPS',
                             'NNS', 'PDT', 'POS', 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO', 'UH', 'VB',
                             'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB', 'PUNCT']
        dict = {k: v for v, k in enumerate(all_possible_tags)}
        X = np.zeros((len(examples), len(all_possible_tags)))
        ii = 0
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            for word_and_pos in words_and_pos_tags:
                tag = word_and_pos[1]
                if tag not in all_possible_tags:
                    tag = 'PUNCT'
                X[ii, dict[tag]] = np.array(len(tag))
            ii = ii + 1
        csr_matrix(X)

class GrammarSentenceTranformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        dict = {}
        for sentence in examples:
            words = word_tokenize(sentence)
            words_and_pos_tags = pos_tag(words)
            for word_and_pos in words_and_pos_tags:
                if word_and_pos[1] in dict:
                    dict[word_and_pos[1]] += 1
                else:
                    dict[word_and_pos[1]] = 1

        X = np.zeros((len(dict), 1));
        i = 0
        for key,value in dict.items():
            X[i, :] = np.array([value])
            i += 1
        return csr_matrix(X)

class BadWordCounter(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.badwords_ = my_badlist

    def get_feature_names(self):
        return np.array(['n_words', 'n_chars', 'allcaps', 'max_len',
                         'mean_len', '@', '!', 'spaces', 'bad_ratio', 'n_bad',
                         'capsratio'])

    def fit(self, examples):
        return self

    def transform(self, examples):
        ## some artificial features!
        n_words = [len(c.split()) for c in examples]
        n_chars = [len(c) for c in examples]
        # number of uppercase words
        allcaps = [np.sum([w.isupper() for w in comment.split()])
                   for comment in examples]
        # longest word
        max_word_len = [np.max([len(w) for w in c.split()]) for c in examples]
        # average word length
        mean_word_len = [np.mean([len(w) for w in c.split()])
                         for c in examples]
        # number of google badwords:
        n_bad = [np.sum([c.lower().count(w) for w in self.badwords_])
                 for c in examples]
        exclamation = [c.count("!") for c in examples]
        addressing = [c.count("@") for c in examples]
        spaces = [c.count(" ") for c in examples]

        allcaps_ratio = np.array(allcaps) / np.array(n_words, dtype=np.float)
        bad_ratio = np.array(n_bad) / np.array(n_words, dtype=np.float)

        return np.array([n_words, n_chars, allcaps, max_word_len,
                         mean_word_len, exclamation, addressing, spaces, bad_ratio, n_bad,
                         allcaps_ratio]).T
    
class RatingMovieTransform(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1))
        ii = 0
        for x in examples:
            X[ii, :] = x
            ii += 1
        return csr_matrix(X)
    
class UserVoteTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, examples):
        return self

    def transform(self, examples):
        X = np.zeros((len(examples), 1))
        ii = 0
        for x in examples:
            X[ii, :] = x
            ii += 1
        return csr_matrix(X)
```

```python
class TextFeatureTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.badwords_ = my_badlist
        self.subjectivity = load_subjectivity()
        self.stemmer = nltk.stem.PorterStemmer()

    def get_feature_names(self):
        feature_names = []
        feature_names.extend(self.unigram_vect.get_feature_names())
        feature_names.extend(self.bigram_vect_you.get_feature_names())
        feature_names.extend(self.trigram_vect_you.get_feature_names())
        feature_names.extend(["you_are_" + w for w in
                              self.you_are_vect.get_feature_names()])
        feature_names.extend(["n_nicks", "n_urls", "n_sentences",
                              "n_non_words", "idiot_regexp", "moron_regexp", "n_html"])
        feature_names.extend(["strong_pos", "strong_neg", "weak_pos",
                              "weak_neg"])
        feature_names.extend(['n_words', 'n_chars', 'toolong', 'allcaps',
                              'max_len', 'mean_len', 'bad_ratio',
                              'n_bad', 'capsratio'])
        feature_names = [" ".join(w) if isinstance(w, tuple) else w
                         for w in feature_names]
        return np.array(feature_names)

    def fit(self, comments, y=None):
        self.fit_transform(comments, y)
        return self

    def fit_transform(self, comments, y=None):
        designed, filtered_words_lower, filtered_words, comments_prep = \
            self.preprocess(comments)

        empty_analyzer = lambda x: x
        self.unigram_vect = TfidfVectorizer(analyzer=empty_analyzer, min_df=3)
        print("comment vecorizing")
        unigrams = self.unigram_vect.fit_transform(filtered_words_lower)
        self.you_are_vect = TfidfVectorizer(
            token_pattern="(?i)you are(?: an?)?(?: the)?(?: as)? (\w+)")
        you_are = self.you_are_vect.fit_transform(comments_prep)
        self.bigram_measures = col.BigramAssocMeasures()
        self.trigram_measures = col.TrigramAssocMeasures()
        # extract bigram collocations including "you" (and your?)
        col_you_bi = col.BigramCollocationFinder.from_documents(
            filtered_words_lower)
        col_you_bi.apply_freq_filter(3)
        col_you_bi._apply_filter(lambda x, y: np.all([w != "you" for w in x]))
        self.you_bigrams = col_you_bi.nbest(self.bigram_measures.chi_sq, 1000)
        self.col_you_bi = col_you_bi
        # make tfidfvectorizer that uses these bigrams
        self.bigram_vect_you = TfidfVectorizer(
            analyzer=make_collocation_analyzer(self.you_bigrams), min_df=3)
        you_bigrams = self.bigram_vect_you.fit_transform(filtered_words_lower)
        # extract trigram collocations
        trigram_col_you = col.TrigramCollocationFinder.from_documents(
            filtered_words_lower)
        trigram_col_you.apply_freq_filter(3)
        trigram_col_you._apply_filter(lambda x, y: np.all([w != "you" for w in x]))
        self.you_trigrams = trigram_col_you.nbest(self.trigram_measures.chi_sq,1000)
        self.trigram_col_you = trigram_col_you
        self.trigram_vect_you = TfidfVectorizer(
            analyzer=make_collocation_analyzer(self.you_trigrams, 3), min_df=3)
        you_trigrams = self.trigram_vect_you.fit_transform(
            filtered_words_lower)
        ## some extra features features!
        designed.extend(self.extra_features(filtered_words, comments,
                                          filtered_words_lower, ))
        designed = np.array(designed).T
        self.scaler = MinMaxScaler()
        designed = self.scaler.fit_transform(designed)
        features = []
        features.append(unigrams)
        features.append(you_bigrams)
        features.append(you_trigrams)
        features.append(you_are)
        # features.append(pos_unigrams)
        features.append(sparse.csr_matrix(designed))
        features = sparse.hstack(features).tocsr()

        return features

    def preprocess(self, comments):
        # remove nicknames, urls, html
        nick = re.compile(r"@\w\w+:?")
        url = re.compile(r"http[^\s]*")
        html = re.compile(r"</?\w+[^>]*>")
        n_html = [len(html.findall(c)) for c in comments]
        comments = [html.sub(' ', c) for c in comments]

        n_nicks = [len(nick.findall(c)) for c in comments]
        comments_nonick = [nick.sub('', c) for c in comments]

        n_urls = [len(url.findall(c)) for c in comments_nonick]
        comments_nourl = [url.sub(' ', c) for c in comments_nonick]
        comments_ascii = [c.replace(u'\xa0', ' ') for c in comments_nourl]
        comments_ascii = [remove_non_ascii(c) for c in comments_ascii]
        comments_ascii = [
            c.replace("'ll", "will").replace("n't", "not")
                .replace("'LL", "WILL").replace("N'T", "NOT")
            for c in comments_ascii]
        # replace /  with space, as this often separates words
        comments_ascii = [c.replace(u'/', ' ') for c in comments_ascii]

        ur = "you are "
        UR = "YOU ARE "
        comments_ascii = [re.sub(r"[Yy]ou'? ?a?re ", ur, c)
                          for c in comments_ascii]
        # again for the loud people (don't want to lose that)
        comments_ascii = [re.sub(r"YOU'? ?A?RE ", UR, c)
                          for c in comments_ascii]
        idiot = [len(re.findall("you.? [\w ]* idi.t", c))
                 for c in comments_ascii]
        moron = [len(re.findall("you.? [\w ]* m.r.n", c))
                 for c in comments_ascii]
        # split into sentences
        sentences = [nltk.sent_tokenize(comment)
                     for comment in comments_ascii]
        # remove dots as they are annoying
        sentences = [[s.replace(".", " ") for s in sent] for sent in
                     sentences]
        # punctuation = \
        # ['...', '.', '?', '!', ',', "''", '``', '#', '$', "'", "%", "&"]
        n_sentences = [len(sent) for sent in sentences]
        words = [[nltk.word_tokenize(s) for s in sent] for sent in sentences]
        flat_words = [[w for sent in sents for w in sent] for sents in words]
        # remove "words" that contain no letter/numbers
        filtered_words = [[w for w in c
                           if not re.findall(r"^[^\w]*$", w)] for c in flat_words]

        # get rid of non-word characters sourrounding words
        filtered_words = [[re.sub("^[^\w]*(\w+)[^\w]*$", r"\1", w) for w in c]
                          for c in filtered_words]
        # laughter normalization ^^
        filtered_words = [[re.sub("(?i)ha(ha)+", r"haha", w) for w in c]
                          for c in filtered_words]
        filtered_words = [[re.sub("(?i)l+o+l+(o+l+)+", r"lol", w) for w in c]
                          for c in filtered_words]
        # replace the famous "0" as o
        filtered_words = [[re.sub("(?i)([a-z]+)0([a-z]+)", r"\1O\2", w)
                           for w in c] for c in filtered_words]

        filtered_words = [[self.stemmer.stem(w)
                           for w in c] for c in filtered_words]

        # detect weird stuff so we can spellcheck
        non_words = [[a for a in s if not nltk.corpus.wordnet.synsets(a)]
                     for s in filtered_words]
        non_words = [[a for a in s if not a.lower() in self.badwords_]
                     for s in non_words]
        n_non_words = [len(w) for w in non_words]
        filtered_words_lower = [[w.lower() for w in comment]
                                for comment in filtered_words]
        features = [n_nicks, n_urls, n_sentences, n_non_words, idiot, moron,
                    n_html]
        return [features, filtered_words_lower,
                filtered_words, comments_ascii]

    def extra_features(self, filtered_words, comments, filtered_words_lower):
        ## some extra features features!
        n_words = np.array([len(c) for c in filtered_words], dtype=np.float)
        n_words += 0.1
        n_chars = [len(c) for c in comments]
        too_long = np.array(n_chars) > 1000
        # number of uppercase words
        allcaps = [np.sum([w.isupper() for sentence in comment
                           for w in sentence])
                   for comment in filtered_words]
        # after removeing all the stuff above, the comment migh be empty
        max_word_len = [np.max([len(w) for w in c])
                        if len(c) else 0 for c in filtered_words]
        # average word length
        mean_word_len = [np.mean([len(w) for w in c])
                         if len(c) else 0 for c in filtered_words]
        n_bad = [np.sum([self.stemmer.stem(w) in self.badwords_
                         for w in c])
                 if len(c) else 0 for c in filtered_words_lower]

        allcaps_ratio = np.array(allcaps) / n_words
        bad_ratio = np.array(n_bad) / n_words
        strong_pos = [np.sum([w in self.subjectivity[0] for w in c])
                      if len(c) else 0 for c in filtered_words_lower]
        strong_pos = np.array(strong_pos) / n_words
        strong_neg = [np.sum([w in self.subjectivity[1] for w in c])
                      if len(c) else 0 for c in filtered_words_lower]
        strong_neg = np.array(strong_pos) / n_words
        weak_pos = [np.sum([w in self.subjectivity[2] for w in c])
                    if len(c) else 0 for c in filtered_words_lower]
        weak_pos = np.array(strong_pos) / n_words
        weak_neg = [np.sum([w in self.subjectivity[3] for w in c])
                    if len(c) else 0 for c in filtered_words_lower]
        weak_neg = np.array(strong_pos) / n_words

        result = [strong_pos, strong_neg, weak_pos, weak_neg, n_words, n_chars,
                  allcaps, too_long, max_word_len, mean_word_len, bad_ratio,
                  n_bad, allcaps_ratio]
        return result

    def transform(self, comments):
        designed, filtered_words_lower, filtered_words, comments_prep = \
            self.preprocess(comments)
        unigrams = self.unigram_vect.transform(filtered_words_lower)
        you_bigrams = self.bigram_vect_you.transform(filtered_words_lower)
        you_trigrams = self.trigram_vect_you.transform(filtered_words_lower)
        you_are = self.extra_features.transform(comments_prep)
        designed.extend(self.extra_features(filtered_words, comments,
                                          filtered_words_lower))
        designed = np.array(designed).T
        designed = self.scaler.transform(designed)
        features = []
        features.append(unigrams)
        features.append(you_bigrams)
        features.append(you_trigrams)
        features.append(you_are)
        features.append(sparse.csr_matrix(designed))
        features = sparse.hstack(features).tocsr()
        return features
```

```python
class FeatEngr:
    def __init__(self):
        self.vectorizer = TfidfVectorizer(analyzer='word', stop_words='english', ngram_range=(1, 2), lowercase=True,
                                          norm='l2', tokenizer=WordStemmerTokenizer(),
                                          token_pattern=r"[\!\"\#\$\%\&\'\(\)\*\+\,\-\.\/\:\;\<\=\>\?\@\[\\\]\^_\`\{\|\}\~]")

    def build_train_features(self, examples):
        self.all_features = FeatureUnion([
            ('text_stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('text_length', TextLengthTransformer())
            ])),
            ('Bag-of-words', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('vectorizer', self.vectorizer)
            ])),
            # trope
            ('trope-hotcoding', Pipeline([
                ('selector', ItemSelector(key=kTROPE_FIELD)),
                ('trope-encoding', TfidfVectorizer(max_df=0.8, lowercase=True, norm='l2'))
            ])),
            ('trope-sentence', Pipeline([
                ('selector', ItemSelector(key=kSENTENCE_TROPE)),
                (
                'trope-sentence-encoding', TfidfVectorizer(stop_words='english', max_df=0.7, lowercase=True, norm='l2'))
            ])),

            ('grammar-adjective-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('grammar-adjective-encoding', GrammarAdjectiveTranformer())
            ])),
            ('grammar-noun-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('grammar-noun-encoding', GrammarNounTranformer())
            ])),
            ('bad-words-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('bad-word-encoding', BadWordCounter())
            ])),
            ('Text-Feature-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('Text-Feature-encoding', TextFeatureTransformer())
            ])),
            ('Generes-Feature-stats', Pipeline([
                ('selector', ItemSelector(key=kGENRE_FIELD)),
                ('Genre-Feature-Encoding', HashingVectorizer(stop_words='english', lowercase=True, ngram_range=(2, 2))),
            ])),
            ('spoiler-word-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('spoiler-word-Encoding', SpoilerWordDetectionTransform()),
            ])),
            ('rating-movie-stats', Pipeline([
                ('selector', ItemSelector(key=kRATING_FIELD)),
                ('rating-movie-Encoding', RatingMovieTransform()),
            ]))

        ])

        '''
            ('feature-stacker', Pipeline([
                ('vect', ft), 
                ('select', select), 
                ('logr', clf)]))
          ('user-votes-stats', Pipeline([
                ('selector', ItemSelector(key=kUSER_VOTE_FIELD)),
                ('user-votes-Encoding', UserVoteTransformer()),
            ]))
         ('page-stats', Pipeline([
                ('selector', ItemSelector(key=kPAGE_FIELD)),
                ('page-encoding', TfidfVectorizer(max_df=0.8, lowercase=True, norm='l2'))
            ])),
        ('pos-sentence-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('pos-sentence-encoding', POSWordAssociationTokenizer())
            ]))
        ('page-sentence-stats', Pipeline([
                ('selector', ItemSelector(key=kPAGE_SENTENCE)),
                ('page-sentence-encoding', TfidfVectorizer(stop_words='english', max_df=0.8, lowercase=True, norm='l2'))
            ])),
            ('page-trope-stats', Pipeline([
                ('selector', ItemSelector(key=kPAGE_TROPE)),
                ('page-trope-encoding', TfidfVectorizer(max_df=0.8, lowercase=True, norm='l2'))
            ])),'''
        '''('page-sentence-trope-stats', Pipeline([
            ('selector', ItemSelector(key=kPAGE_SENTENCE_TROPE)),
            ('page-sentence-trope-encoding', TfidfVectorizer(stop_words='english', max_df=0.8, lowercase=True, norm='l2'))
        ])),

            ('POS-Word-Count-stats', Pipeline([
                ('selector', ItemSelector(key=kTEXT_FIELD)),
                ('POS-Word-Count-encoding', POSWordCountTransformer())
            ]))
        
        '''
        return self.all_features.fit_transform(examples)

    def get_test_features(self, examples):
        Y = self.all_features.transform(examples)
        return Y

    def error_analysis(self,error_analysis_list, prediction, actual_prediction):
        if len(prediction) != len(actual_prediction):
            print('Error!!! Both prediction and actual predictions are not of same size')
            return;
        print(classification_report(actual_prediction, prediction,))
        p = 0
        q = len(prediction)
        for i in range(0, q):
            if actual_prediction[i] != prediction[i]:
                p += 1

        error_analysis_list.append((float(p)/q)*100)
            
        
    def show_top10(self):
        feature_names = np.asarray(self.vectorizer.get_feature_names())
        top10 = np.argsort(self.logreg.coef_[0])[-10:]
        bottom10 = np.argsort(self.logreg.coef_[0])[:10]
        print("Pos: %s" % " ".join(feature_names[top10]))
        print("Neg: %s" % " ".join(feature_names[bottom10]))

    def train_model(self, random_state=1234):

        import pickle
        # load data
        # dfTrain = pd.read_csv("data/spoilers/train.csv")
        dfTrain = train_file_load
        dump = movie_rating_file_load
        train_rating = dump[0]
        page = dfTrain[kPAGE_FIELD]
        rating_list = []
        for p in page:
            if p in train_rating:
                rating_list.append(int(float(train_rating[p]) * 10))

        dfTrain[kRATING_FIELD] = rating_list
        dump = movie_user_rating_file_load
        train_user_voting = dump[0]
        page = dfTrain[kPAGE_FIELD]
        user_voting_list = []
        for p in page:
            if p in train_user_voting:
                user_voting_list.append(sum(train_user_voting[p].values()))
        dfTrain[kUSER_VOTE_FIELD] = user_voting_list
        dfTrain[kSENTENCE_TROPE] = dfTrain[[kTEXT_FIELD, kTROPE_FIELD]].apply(lambda x: ' '.join(x), axis=1)
        dfTrain[kPAGE_SENTENCE] = dfTrain[[kTEXT_FIELD, kPAGE_FIELD]].apply(lambda x: ' '.join(x), axis=1)
        dfTrain[kPAGE_TROPE] = dfTrain[[kPAGE_FIELD, kTROPE_FIELD]].apply(lambda x: ' '.join(x), axis=1)
        dfTrain[kPAGE_SENTENCE_TROPE] = dfTrain[[kTEXT_FIELD, kPAGE_FIELD, kTROPE_FIELD]].apply(lambda x: ' '.join(x),
                                                                                                axis=1)

        self.X_train = self.build_train_features(dfTrain)
        self.y_train = np.array(dfTrain[kTARGET_FIELD], dtype=int)
        from sklearn.model_selection import KFold, StratifiedKFold, ShuffleSplit
        average_accuracy = 0
        count = 0
        kf = KFold(n_splits=10, random_state=None, shuffle=True)
        #kf = ShuffleSplit(n_splits=10, random_state=0, test_size=.25)
        error_analysis_list = []
        for train_index, test_index in kf.split(self.X_train):
            loc_X_train, loc_X_test = self.X_train[train_index], self.X_train[test_index]
            loc_y_train, loc_y_test = self.y_train[train_index], self.y_train[test_index]

            # train logistic regression model.  !!You MAY NOT CHANGE THIS!!
            self.logreg = LogisticRegression(random_state=random_state)
            self.logreg.fit(loc_X_train, loc_y_train)
            loc_y_pred = self.logreg.predict(loc_X_test)
            local_accuracy = accuracy_score(loc_y_test, loc_y_pred)
            self.error_analysis(error_analysis_list, loc_y_pred, loc_y_test)
            print("local accuracy = ", local_accuracy, " error rate ----->", error_analysis_list[count])
            print("-------------------------------------------------------------------------------------")
            average_accuracy += local_accuracy
            count += 1

        print("average accuracy = ", (average_accuracy / count) * 100)
        print("average error rate = ", (sum(error_analysis_list)/count))

    def model_predict(self):
        # read in test data
        import pickle
        # dfTest = pd.read_csv("data/spoilers/test.csv")
        dfTest = test_file_load

        dump = movie_rating_file_load
        train_rating = dump[1]
        page = dfTest[kPAGE_FIELD]
        rating_list = []
        for p in page:
            if p in train_rating:
                rating_list.append(int(train_rating[p]) * 10)

        dfTest[kRATING_FIELD] = rating_list
        dump = movie_user_rating_file_load
        train_user_voting = dump[1]
        page = dfTest[kPAGE_FIELD]
        user_voting_list = []
        for p in page:
            if p in train_user_voting:
                user_voting_list.append(sum(train_user_voting[p].values()))
        dfTest[kUSER_VOTE_FIELD] = user_voting_list
        dfTest[kSENTENCE_TROPE] = dfTest[[kTEXT_FIELD, kTROPE_FIELD]].apply(lambda x: ' '.join(x), axis=1)
        dfTest[kPAGE_SENTENCE] = dfTest[[kTEXT_FIELD, kPAGE_FIELD]].apply(lambda x: ' '.join(x), axis=1)
        dfTest[kPAGE_TROPE] = dfTest[[kPAGE_FIELD, kTROPE_FIELD]].apply(lambda x: ' '.join(x), axis=1)
        dfTest[kPAGE_SENTENCE_TROPE] = dfTest[[kTEXT_FIELD, kPAGE_FIELD, kTROPE_FIELD]].apply(lambda x: ' '.join(x),
                                                                                              axis=1)
        self.X_test = self.get_test_features(dfTest)
        # make predictions on test data
        pred = self.logreg.predict(self.X_test)

        # dump predictions to file for submission to Kaggle
        pd.DataFrame({kTARGET_FIELD: np.array(pred, dtype=bool)}).to_csv("prediction.csv", index=True, index_label="Id")
```

### Feature 5  : POS Tag Encoding

```python
import pandas as pd
from matplotlib import pyplot as plt
import nltk

all_possible_tags = ['WRB', 'WP$', 'WP', 'WDT', 'VBZ', 'VBP', 'VBN', 'VBG', 'VBD', 'VB', 'UH', 'TO', 'SYM', 'RP', 'RBS', 'RBR', 'RB', 'PRP$', 'PRP', 'POS', 'PDT', 'NNS', 'NNPS', 'NNP', 'NN', 'MD', 'LS', 'JJS', 'JJR', 'JJ', 'IN', 'FW', 'EX', 'DT', 'CD', 'CC']
all_possible_tags = sorted(all_possible_tags)
spoiler_pos_dict = {}
non_spoiler_pos_dict = {}

for tag in all_possible_tags:
    spoiler_pos_dict[tag] = 0
    non_spoiler_pos_dict[tag] = 0

dfTrain = pd.read_csv("../data/spoilers/train.csv")

for index,row in dfTrain.iterrows():
    sentence = row[kTEXT_FIELD]
    tags = nltk.pos_tag(sentence)

    spoiler = row[kTARGET_FIELD]
    d = None
    if spoiler:
        d = spoiler_pos_dict
    else:
        d = non_spoiler_pos_dict

    for tup in tags:
        if tup[-1] in d.keys():
            d[tup[-1]] += 1
        else:
            pass

print(spoiler_pos_dict)
print(non_spoiler_pos_dict)

D = spoiler_pos_dict

x = range(len(D))

y = []
z = []
for key,value in D.items():
    y.append(value)
    z.append(key)
plt.figure(figsize=(20,6))
plt.bar(x, y, align='center',color = 'red')
plt.xlabel('Part of Speech Tag')
plt.ylabel('Count of words')
plt.title('Spoiler : POS word Count')
plt.xticks(x, z)
plt.show()


y = []
z = []
D = non_spoiler_pos_dict
x = range(len(D))
for key,value in D.items():
    y.append(value)
    z.append(key)

plt.figure(figsize=(20,6))
plt.bar(x, y, align='center',color = 'black')
plt.xlabel('Part of Speech Tag')
plt.ylabel('Count of words')
plt.title('Non-Spoiler : POS word Count')
plt.xticks(x, z)

plt.show()
```

From the above graph, It's clear that pos tag is not playing much role as for the spoiler and non-spoiler graph, we are getting the almost same graph. So, this feature is not helping much.

### Feature 6 : Rating Stats

```python
from matplotlib import pyplot as plt


dfTrain = train_file_load
dump = movie_rating_file_load
train_rating = dump[0]
page = dfTrain[kPAGE_FIELD]
rating_list = []
for p in page:
    if p in train_rating:
        rating_list.append(train_rating[p])

dfTrain[kRATING_FIELD] = rating_list
spoiler_list = dfTrain.loc[dfTrain[kTARGET_FIELD] == 1]
non_spoiler_list = dfTrain.loc[dfTrain[kTARGET_FIELD] == 0]


page_group_spoiler = spoiler_list.groupby(kRATING_FIELD).size().sort_values(ascending=False)
page_group_nonspoiler = non_spoiler_list.groupby(kRATING_FIELD).size().sort_values(ascending=False)


fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(page_group_spoiler.index, page_group_spoiler.values, s=10, c='r', marker="s", label='spoiler')
ax1.scatter(page_group_nonspoiler.index, page_group_nonspoiler.values, s=10, c='b', marker="o", label='non-spoiler')
plt.xlabel('Rating')
plt.ylabel('count of pages')
plt.legend(loc='upper left')
plt.title('Rating : Spoiler and non-spoiler Analysis')
plt.show()
```
