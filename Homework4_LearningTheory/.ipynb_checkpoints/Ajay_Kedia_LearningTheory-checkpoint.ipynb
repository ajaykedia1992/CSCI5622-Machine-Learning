Consider the class C of concepts defined by triangles with **distinct** vertices of the form $(i, j)$ where $i$ and $j$ are integers in the interval $[0,99]$. A concept c labels points on the interior and boundary of the associated triangle as positive and points exterior to the triangle as negative.

**Note**: To make life easier, we'll allow degenerate triangles in $C$. That is, triangles where the vertices are collinear. The following image depicts an example of a nondegenerate and a degenerate triangle.

**Part A**: Suppose we have an algorithm that produces a consistent $h$ from the hypothesis class $H = C$. Give a bound on the number of training examples sufficient to assure that for any target concept $c$ in $C$, our algorithm will, with probability $1-\delta$, output a hypothesis $h$ with generalization error at most $\epsilon$.

### Answer A

\begin{align*}
	m &\geq \frac{1}{\epsilon}\bigg(\ln \vert H \vert + \ln\frac{1}{\delta}\bigg)\\
	\\
	\delta &= 1-0.95 = 0.05\\
	\epsilon &= 0.15\\
	\\
	\vert H \vert &= \text{number of possible triangles in the space}\\
	\vert H \vert &= {{\text{possible vertices}}\choose{3}}\\
    \text{as the triangle is collinear and degenerated}\\
	\text{possible vertices} &= 100^2 = 10000\\
	\vert H \vert &= {{10000}\choose{3}}=166616670000\\
    \ln \vert H \vert = 25.8389\\
    m &\geq \frac{1}{\epsilon}\bigg(25.8389 + \ln\frac{1}{\delta}\bigg)\\
	\\
\end{align*}

**Part B**: Based on your bound in **Part A**, determine the minimum number of training examples necessary such that for any target concept $c$ in $C$, our algorithm will, with probability $0.95$, output a hypothesis $h$ with generalization error at most $0.15$.  

### Answer B

\begin{align*}
	m &\geq \frac{1}{\epsilon}\bigg(25.8389 + \ln\frac{1}{\delta}\bigg)\\
	\\
	\delta &= 1-0.95 = 0.05\\
	\epsilon &= 0.15\\
	m &\geq \frac{1}{.15}\bigg(25.8389  + \ln\frac{1}{0.05}\bigg)\\
    m = 192.36\\
	m = ceil \bigg(192.36 \bigg)\\
    m = 193
\end{align*}

### [15 points] Problem 2: 
***

Consider feature vectors that live in two-dimensional space and the class of hypotheses defined by circles **centered at the origin**. There are two different kinds of hypotheses $h$ in this class. One type of hypthesis classifies points as positive if they lie on the boundary or **interior** of the circle, and negative otherwise. The other type of hypothesis classifies points as positive if they lie on the boundary or **exterior** of the circle, and negative otherwise. State and prove (rigorously) the VC dimension of this family of classifiers.

### Answer 2
The VC dimension for origin-centered circles is 2.
Consider a set of 2 points with radii r1 < r2 from the origin. There exists a hypothesis 'h' that can label the points at r1 positive while r2 is labeled negative or vice versa. A hypothesis also exists that can label both points positive or both points negative. So there exists a set of 2 points that can be shattered by an origin-centered circle. 
With a set of 3 points at radii r1<=r2<=r3 from the origin no hypothesis 'h' will be able to label r1 and r3 as positive while also labeling r2 negative. Thus, no set of three points can be shatered by an origin-centered circle.




\begin{align*}
    H = SIGN \bigg(X^T X - \Theta \bigg)\\
\end{align*}

We can say a classifier f(x) can shatter points X1, X2, ... Xn iff for all Y1, Y2, ... Yn, f(x) can achieve zero error on training data(X1,Y1), (X2,Y2), ... (Xn,Yn). So there exist some \Theta which can achieve zero error on our training data.

### [20 points] Problem 3: Empirical Verification of PAC Bounds for Axis-Aligned Rectangles 
***

In the in-class notebook associated with PAC Learnability, we proved a PAC bound for the class of concepts $C$ comprised of axis-aligned rectangles living in $\mathbb{R}^2$ of the form $(a \leq x \leq b) \wedge (c \leq y \leq d)$ where $a, b, c, d$ are real numbers. Specifically, we proved that with probability $1-\delta$, any consistent learner could learn a hypothesis $h$ in $H = C$ with generalization error less than $\epsilon$ provided that the number of training examples satisfied 

$$
m > \frac{4}{\epsilon}\log\frac{4}{\delta}
$$

In this problem you will empirically verify this bound for the restricted concept class $C$ where the rectangles are defined by $(a \leq x \leq b) \wedge (c \leq y \leq d)$ where $a, b, c, d$ are real numbers satisfying $0 \leq a \leq b \leq 100$ and $0 \leq c \leq d \leq 100$.

- Write some code that, given training examples of length-2, and labeled according to a concept $c$, returns a consistent hypothesis rectangle $h$. 



- Write some code that generates a training set of size $m$, labels them according to a random concept $c$, learns a consistent hypothesis $h$, and then approximates the generalization error by predicting on $1000$ new examples from the same distribution as the training data. 


- Write some code that computes approximate generalization errors for $100$ independent concepts $c$ and associated training sets of size $m$, and returns the worst-case generalization error at the confidence level $1-\delta$.  One way to do this in the case that say $\delta = 0.05$, is to report the $95^\textrm{th}$ percentile of the $100$ samples of the generalization error. We can then say that, in our simulation, $100(1-\delta)\%$ of our observed generalization errors were less than our computed value. (**Bonus**: If your code is efficient, try increasing the number of runs in the simulation to $500$. This should give you a better approximation of the generalization error.) 

```python
import time
import math
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as patches

def generate_rectangle_axis(a,b):
    s = np.random.uniform(a,b,2)
    s = np.sort(s)
    x1 = s[0]
    y1 = s[1]
    return x1,y1

def generate_random_features(x,y,p,q): 
    min_x = x
    max_x = x + width
    min_y = y
    max_y = y + height
    features = []
    hypothesis_feature = []
    for i in range(len(p)):
        col = "orange"
        if p[i]>=min_x - 0.5 and p[i]<=max_x+0.5 and q[i]>=min_y - 0.5 and q[i]<=max_y + 0.5:
            col = "blue"
            hypothesis_feature.append((p[i],q[i]))
        tup = (p[i],q[i],col)
        features.append(tup)
        
    return features,hypothesis_feature    
    
def get_key_x(tup):
    return tup[0]

def get_key_y(tup):
    return tup[1]

def generate_hypothesis(feature_x,feature_y):
    min_x = feature_x[0][0]
    max_x = feature_x[len(feature_x) - 1][0]
    min_y = feature_y[0][1]
    max_y = feature_y[len(feature_y) - 1][1]
    
    width_h = max_x - min_x
    height_h = max_y - min_y
    return min_x,min_y,width_h,height_h

def hypothesis_miss_count(hypothesis_feature,x_h,y_h, width_h,height_h):
    count = 0
    for i in range(len(hypothesis_feature)):
        x = hypothesis_feature[i][0]
        y = hypothesis_feature[i][1]
        if x<x_h or x>x_h + width_h or y<y_h or y>y_h+height_h:
            count += 1
    return count
```

```python
width = 60
height = 50
a = 1
b = 100
total_run = 500
m = 1000
generalization_error_list = []
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
x,y = generate_rectangle_axis(10,40)
axes.add_patch(patches.Rectangle((x, y),width,height, color = 'darkblue', alpha = 1)) 

p = np.random.uniform(a,b,100)
q = np.random.uniform(a,b,100)
features,hypothesis_feature = generate_random_features(x,y,p,q)
for i  in range(len(features)):
    tup = features[i]
    plt.plot(tup[0], tup[1], "o", color = tup[2])

feature_x_sort = sorted(hypothesis_feature, key = get_key_x)
feature_y_sort = sorted(hypothesis_feature, key = get_key_y)

x_h, y_h, width_h, height_h = generate_hypothesis(feature_x_sort,feature_y_sort)

axes.add_patch(patches.Rectangle((x_h, y_h),width_h,height_h, color = 'sandybrown', alpha = 1))



p = np.random.uniform(a,b,m)
q = np.random.uniform(a,b,m)


features,hypothesis_feature = generate_random_features(x,y,p,q)
for i  in range(len(features)):
    tup = features[i]
    plt.plot(tup[0], tup[1], "o", color = tup[2])

miss_count = hypothesis_miss_count(hypothesis_feature,x_h,y_h, width_h,height_h)

generalization_error = miss_count/m
print("Miss Count = ", miss_count)
print("Generalization error = ", generalization_error)
red_patch = patches.Patch(color='orange', label='Negative Data')
blue_patch = patches.Patch(color='blue', label='positive Data')
plt.legend(handles=[red_patch,blue_patch])
plt.xlabel("X range of features")
plt.ylabel("Y range of features")
axes.set_xlim([0,100]); axes.set_ylim([0,100]); 
```

```python
width = 60
height = 50
a = 1
b = 100
total_run = 500

def estimate_generalization_error(m):
    generalization_error_list = []
    for i in range(total_run):
        x,y = generate_rectangle_axis(10,40)
        p = np.random.uniform(a,b,m)
        q = np.random.uniform(a,b,m)
        features,hypothesis_feature = generate_random_features(x,y,p,q)
        feature_x_sort = sorted(hypothesis_feature, key = get_key_x)
        feature_y_sort = sorted(hypothesis_feature, key = get_key_y)
        x_h, y_h, width_h, height_h = generate_hypothesis(feature_x_sort,feature_y_sort)

        p = np.random.uniform(a,b,1000)
        q = np.random.uniform(a,b,1000)

        features,hypothesis_feature = generate_random_features(x,y,p,q)
        miss_count = hypothesis_miss_count(hypothesis_feature,x_h,y_h, width_h,height_h)

        generalization_error = miss_count/1000
        generalization_error_list.append(generalization_error)
        return generalization_error_list
    
generalization_error_list = estimate_generalization_error(100)    
generalization_error_list = np.asarray(generalization_error_list)
print("Overall generalization error for 500 concept classes at 95 percentile = ", np.percentile(generalization_error_list,95))
```

The statement that $P[R(h) > \epsilon]$ is equivalent to the probability that no point in the training set fell into the missed region.  Thus, by the Union Bound, we have 

$$
\begin{eqnarray}
P[R(h) > \epsilon] &=& P[\textrm{no } {\bf x}_i \textrm{ fell in } r_1 \textrm{ OR } r_2 \textrm{ OR } r_3 \textrm{ OR } r_4]
\end{eqnarray}
$$

By construction, the probability that a single point did not fall in a particular region is $(1-\epsilon/4)$, and thus the probability that none of the $m$ points fell in a particular region is $(1-\epsilon/4)^m$.  Using the Union bound, and these probabilities, we have 

$$
\begin{eqnarray}
P[R(h) > \epsilon] &=& P[\textrm{no } {\bf x}_i \textrm{ fell in } r_1 \textrm{ OR } r_2 \textrm{ OR } r_3 \textrm{ OR } r_4] \\
&\leq& \left(1 - \frac{\epsilon}{4}\right)^m + \left(1 - \frac{\epsilon}{4}\right)^m + \left(1 - \frac{\epsilon}{4}\right)^m + \left(1 - \frac{\epsilon}{4}\right)^m  \\
&=& 4\left(1 - \frac{\epsilon}{4}\right)^m \\
&\leq& 4e^{-m\epsilon/4}
\end{eqnarray}
$$



$$
4e^{-m\epsilon/4} < \delta \quad \Rightarrow \quad \frac{-m\epsilon}{4} < \ln \frac{\delta}{4} \quad \Rightarrow \quad m > \frac{4}{\epsilon}\ln\frac{4}{\delta}
$$


Thus, for any $\epsilon > 0$ and $\delta > 0$ choosing $m$ greater than $(4/\epsilon)\ln(\delta/4)$ ensures that $P[R(h) \leq \epsilon] \geq 1 - \delta$ 

So, considering this $$  m > \frac{4}{\epsilon}\ln\frac{4}{\delta} $$

if we consider our epsilon error, 0.005 and confidence 0.05 then
PAC bound 
$$
m > \frac{4}{0.005}\ln\frac{4}{0.05}\\
m = 1522.47\\
m = Ceil \bigg(1522.47 \bigg)\\
m \geq 1523
$$

So, in above table, at 1500 features, we have generalization error = 0.001, and if we increase our dataset by 23, we will get consistent hypothesis where 
$$
H = C
$$
